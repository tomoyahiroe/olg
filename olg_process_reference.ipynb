## Settingのための関数

```python
# Maliarグリッド

@njit
def maliar_grid(a_min, a_max, N, theta):
		a_grid = np.empty(N)
		for i in range(1,N+1):
				a_grid[i-1] = a_min + (a_max - a_min)*((i-1)/(N-1))**theta

		return a_grid
```

```python
# a_grid_sdからa_gridへ補間する際のindexと重みを計算
# 事前に補間マップを計算しておくことでループ内で高速化できる

@njit
def build_interp_map(a_grid_sd, a_grid):
    """
    細かいグリッド値（a_grid_sd）を、粗いグリッド（a_grid）の
    隣接 2 点に写像して線形補間するための
    (idx1, idx2, weight1, weight2) を返す。

    戻り値:
        idx1, idx2 … int64 配列（各点が属する index）
        w1,  w2   … float 配列（線形補間の重み: w1 + w2 = 1）
    """

    # ―――― 線形補間マップ用のワーク配列を確保 ――――
    idx1 = np.empty(a_grid_sd.size, dtype=np.int64)  # 各点が属する「下側」グリッド index
    idx2 = np.empty(a_grid_sd.size, dtype=np.int64)  # 各点が属する「上側」グリッド index
    w1   = np.empty(a_grid_sd.size)                  # 線形補間の重み（下側）  ⇒ w1 + w2 = 1
    w2   = np.empty(a_grid_sd.size)                  # 線形補間の重み（上側）


    for i in range(a_grid_sd.size):
        x = a_grid_sd[i]

        # 右端はグリッド境界に張り付ける
        if x >= a_grid[-1]:
            idx1[i] = idx2[i] = a_grid.size - 1
            w1[i], w2[i] = 1.0, 0.0
            continue

        else:
            # どの区間 [k,k+1] に入るか
            k = np.searchsorted(a_grid, x, side="right") - 1
            idx1[i], idx2[i] = k, k + 1

            # 線形補間の重み
            d = (x - a_grid[k]) / (a_grid[k + 1] - a_grid[k])
            w1[i] = 1.0 - d
            w2[i] = d

    return idx1, idx2, w1, w2
```

```python
# 効用関数
@njit
def utility(c, gamma):
    return np.log(c) if gamma == 1.0 else c ** (1 - gamma) / (1 - gamma)

@njit
def mutility(c, gamma):
    return 1.0 / c if gamma == 1.0 else c ** (-gamma)
```

## Setting

```python
class Setting:

    def __init__(self,
                 alpha = 0.40,
                 beta = 0.98,
                 gamma = 1.0,
                 delta = 0.08,
                 b = 0.0,             # 借入制約（北尾先生コードではそもそも設定されていない）

                 # 資産グリッド
                 na = 101,
                 na_sd = 2001,
                 a_min = 0.0,
                 a_max = 25.0,
                 theta = 1.20,

                 psi0 = 0.50,          # 所得代替率（最初）
                 psi1 = 0.25,          # 所得代替率（最後）

                # ライフサイクル
                 Nj = 61,               # 年齢 20–80
                 NjW = 45,              # 就業期間

                # スキルに関する個人の異質性
                 nz = 2,
                 dif = 0.30,          # スキル値設定  array([0.7, 1.3])
                 Pz_high = 0.80,      # 遷移確率行列の対角要素  array([[0.8, 0.2], [0.2, 0.8]])

                # 定常状態計算
                 tol_K_SS = 1e-3,     # 許容誤差
                 max_iter_SS = 2000,  # 最大反復回数
                 adjK_SS = 0.20,      # 更新幅

                #移行過程計算
                 tol_K_TR = 1e-4,
                 max_iter_TR = 50,
                 adjK_TR = 0.05,

                 NT = 100,              # 移行期間
                 TT = 25,               # 所得代替率が psi0 から psi1 に変化するまでの期間
                 ):

        self.alpha = alpha
        self.beta = beta
        self.gamma = gamma
        self.delta = delta
        self.b = b

        self.psi0 = psi0
        self.psi1 = psi1

        self.Nj = Nj
        self.NjW = NjW

        self.nz = nz
        self.dif = dif
        self.Pz_high = Pz_high

        self.na = na
        self.na_sd = na_sd
        self.a_min = a_min
        self.a_max = a_max
        self.theta = theta

        self.NT = NT
        self.TT = TT
        self.tol_K_SS = tol_K_SS
        self.tol_K_TR = tol_K_TR
        self.max_iter_SS = max_iter_SS
        self.max_iter_TR = max_iter_TR
        self.adjK_SS = adjK_SS
        self.adjK_TR = adjK_TR

        # 資産グリッド
        self.a_grid  = maliar_grid(self.a_min, self.a_max, self.na, self.theta)
        self.a_grid_sd = maliar_grid(self.a_min, self.a_max, self.na_sd, self.theta)

        # a_grid_sdからa_gridへ補間する際のindexと重みを計算
        self.a_idx1, self.a_idx2, self.a_w1, self.a_w2 = build_interp_map(self.a_grid_sd, self.a_grid)

        # 個人のスキル
        self.z_grid = np.array([1 - self.dif, 1 + self.dif])
        self.Pz    = np.array([[self.Pz_high, 1 - self.Pz_high],
                               [1 - self.Pz_high, self.Pz_high]])

        # 年齢ごとの人口動態と労働供給
        self.sd_age = np.ones(self.Nj) / self.Nj
        self.Lbar   = self.sd_age[: self.NjW].sum()

        # 所得代替率・年金保険料率パス
        self.psi_path = np.concatenate([
            np.linspace(self.psi0, self.psi1, self.TT),
            np.full(self.NT - self.TT, self.psi1),
        ])
        self.tau_path = self.psi_path * self.sd_age[self.NjW :].sum() / self.sd_age[: self.NjW].sum()
```

# 定常状態

## 定常状態計算のための関数

```python
@njit
def bellman_backward(is_SS:bool, v_next: np.ndarray, r: float, w: float, tau: float, psi: float,
                     beta: float, gamma:float, b:float,
                     a_grid: np.ndarray, a_grid_sd: np.ndarray, na: int,
                     a_idx1: np.ndarray, a_idx2: np.ndarray,
                     a_w1: np.ndarray, a_w2: np.ndarray,
                     z_grid: np.ndarray, Pz: np.ndarray,
                     Nj: int, NjW: int, nz: int):

    """
    ベルマン方程式の backward iteration（前期の Nj to 1 歳を計算）
    定常状態計算では V_t(j+1,z,a) から V_t(j,z,a) を計算
    移行過程計算では V_{t+1}(j+1,z,a) から V_t(j,z,a) を計算

    返り値:
        v_curr      … 前期の価値関数 V_t(j,z,a)
        a_prime_idx … 最適貯蓄インデックス （a_grid_sd 上）
        a_prime     … 最適貯蓄額 (a_grid_sd 上)
    """

    v_curr   = np.empty_like(v_next)
    a_prime_idx  = np.zeros((Nj, nz, na), dtype=np.int64)
    a_prime    = np.zeros_like(v_next)

    # 所得 y_t(j,z) を前もって作成
    yvec = np.empty((Nj, nz))
    for j in range(NjW):
        for e in range(nz):
            yvec[j, e] = (1.0 - tau) * w * z_grid[e]
    yvec[NjW :, :] = psi * w

    # t 期の最終年齢（j = Nj） は貯蓄ゼロで解析解
    for z in range(nz):
        for a in range(na):
            c = yvec[-1, z] + (1.0 + r) * a_grid[a]
            v_curr[-1, z, a] = utility(c, gamma)
            a_prime_idx[-1, z, a] = 0
            a_prime[-1, z, a]  = a_grid_sd[0]

    # Nj-1 to 1 で t 期の政策関数と価値関数を導出
    for j in range(Nj - 2, -1, -1):     # j = Nj-2, …, 0
        for z in range(nz):
            y = yvec[j, z]
            for a in range(na):
                best_v = -1e6
                best_acc = 0
                for acc in range(a_grid_sd.size):
                    c = y + (1.0 + r) * a_grid[a] - a_grid_sd[acc] + b
                    if c <= 0:
                        break  # このa_primeの値で所得より貯蓄が大きい → 以降のforループでも所得より貯蓄が大きい
                               # → break することでこのa_prime以降の価値関数の値はすべて-1e6となり、最適な政策関数、価値関数として選ばれない（これより小さいa_primeの中から選ばれる）
                    EV = 0.0
                    for z_prime in range(nz):
                        if is_SS:
                            Vp = (a_w1[acc] * v_curr[j + 1, z_prime, a_idx1[acc]] +
                                  a_w2[acc] * v_curr[j + 1, z_prime, a_idx2[acc]])
                        else:
                            Vp = (a_w1[acc] * v_next[j + 1, z_prime, a_idx1[acc]] +
                                  a_w2[acc] * v_next[j + 1, z_prime, a_idx2[acc]])
                        EV += Pz[z, z_prime] * Vp
                    val = utility(c, gamma) + beta * EV
                    if val > best_v:
                        best_v, best_acc = val, acc
                v_curr[j, z, a] = best_v
                a_prime_idx[j, z, a] = best_acc
                a_prime[j, z, a]   = a_grid_sd[best_acc]
    return v_curr, a_prime_idx, a_prime
```

```python
@njit
def forward_measures(sd_prev: np.ndarray, a_prime_idx: np.ndarray,
                     a_idx1: np.ndarray, a_idx2: np.ndarray,
                     a_w1: np.ndarray, a_w2: np.ndarray,
                     Pz: np.ndarray, sd_age: np.ndarray, Nj: int, nz: int, na: int):
    """
    1 期だけ人口分布 (μ) を前進（次期の i to Nj 歳を計算）
    定常状態計算では μ_t(j,z,a) からμ_t(j+1,z,a) を計算
    移行過程計算では μ_t(j,z,a) から μ_{t+1}(j+1,z,a) を計算

    返り値:
        mea_next … 次期の人口分布 μ_{t+1}(j,z,a)
    """

    sd_next = np.zeros_like(sd_prev)
    # sd_next[0, :, 0] = 1.0 / (Nj * nz)  # j=1つまり20歳の新世代は均一に生まれる
    sd_next[0, :, 0] = sd_age[0] / nz  # j=1つまり20歳の新世代は均一に生まれる

    for j in range(Nj - 1):             # j = 0..Nj-2
        for z in range(nz):
            for a in range(na):
                m = sd_prev[j, z, a]
                if m == 0.0:
                    continue            # スキップで高速化

                # 資産グリッド補間
                acc = a_prime_idx[j, z, a]          # 最適貯蓄 a′ の細グリッド index
                ai1, ai2 = a_idx1[acc], a_idx2[acc] # 粗グリッド左右
                w1, w2   = a_w1[acc], a_w2[acc]     # 線形補間重み

                # スキル遷移 z → z'
                for z_prime in range(nz):
                    sd_next[j + 1, z_prime, ai1] += m * Pz[z, z_prime] * w1
                    sd_next[j + 1, z_prime, ai2] += m * Pz[z, z_prime] * w2

    # もし資産上限セルに質量が溜まり過ぎたら警告
    mass_maxA = sd_next[:, :, -1].sum()
    if mass_maxA > 1e-3:
        print("mass at maxA =", mass_maxA)
    return sd_next
```

```python
@njit
def aggregate_capital(a_prime: np.ndarray, sd: np.ndarray) -> float:
    return (a_prime * sd).sum()
```

```python
def solve_SS(st:Setting, psi: float, tau: float, K_guess: float):

        """
        返り値
        -------
        K        : 収束した定常状態資本ストック
        ss_dict  : {V, a_prime_idx, a_prime, sd, r, w} を詰めた辞書
        """
        K = K_guess
        for it in range(st.max_iter_SS):

            # 与えられた資本 K から要素価格 (r,w) を計算
            r = st.alpha * (K / st.Lbar) ** (st.alpha - 1) - st.delta
            w = (1 - st.alpha) * (K / st.Lbar) ** st.alpha
            v_next = np.zeros((st.Nj, st.nz, st.na))

            # Nj to 1　歳
            v_curr, a_prime_idx, a_prime = bellman_backward(True,
                v_next, r, w, tau, psi, st.beta, st.gamma, st.b,
                st.a_grid, st.a_grid_sd, st.na, st.a_idx1, st.a_idx2, st.a_w1, st.a_w2,
                st.z_grid, st.Pz, st.Nj, st.NjW, st.nz,
            )

            sd = np.zeros_like(v_next)
            sd[0, :, 0] = 1.0 / (st.Nj * st.nz) # 新世代を均等に投入

            # 1 to Nj 歳
            for _ in range(st.Nj - 1):
                sd = forward_measures(sd, a_prime_idx, st.a_idx1, st.a_idx2, st.a_w1, st.a_w2, st.Pz, st.sd_age, st.Nj, st.nz, st.na)
            K_new = aggregate_capital(a_prime, sd)

            # 誤差確認
            print(f"Iter {it:2d} | K error = {np.max(np.abs(K_new - K)):1.2e}")
            if abs(K_new - K) < st.tol_K_SS:
                break
            K += st.adjK_SS * (K_new - K)

        return K, dict(V=v_curr, a_prime_idx=a_prime_idx, a_prime=a_prime, sd=sd, r=r, w=w)
```

```python
def plot_ss(K, ss, st, tau, *, jc_plot: int = 21) -> None:
    """
    MATLAB 版で描いていた 3 枚のグラフ
        ① 資産ストック（年齢×スキル別，消費=1 で規格化）
        ② 貯蓄率   （年齢 = jc_plot でスキル別，資産グリッド上）
        ③ 貯蓄水準 （　　 〃　　　　　　　　　　　　　　 ）
    をまとめて描画。

    Parameters
    ----------
    K  : float
        定常状態で収束した資本ストック
    ss : dict
        `solve_SS()` が返す辞書。
            - "sd"         (μ(j,z,a) : 人口分布)
            - "a_prime"    (a′(j,z,a): 最適次期資産)
            - "r", "w"     (要素価格)
    st : Setting
        モデル諸元を保持する `Setting` インスタンス。
    tau : float
        所得税率
    jc_plot : int, optional
        貯蓄率・貯蓄水準を描く年齢インデックス（1-始まり）。
        デフォルト 21 ⇒ 実年齢 40 歳。

    """

    # ────────── 取り出し ──────────
    a_grid = st.a_grid
    sd     = ss["sd"]
    a_pr   = ss["a_prime"]
    r, w   = ss["r"], ss["w"]

    Nj, NjW, nz, na = st.Nj, st.NjW, st.nz, st.na

    # 年齢×スキル平均資産
    mass_age_skill = sd.sum(axis=2)                              # 年齢j、スキルz別の総人口
    asset_totals   = (a_grid[None, None, :] * sd).sum(axis=2)    # 年齢j、スキルz別の総資産
    afunJE = asset_totals / mass_age_skill                       # 年齢j、スキルz別の平均資産

    # 消費・貯蓄・貯蓄率
    yvec = np.empty((Nj, nz))
    yvec[:NjW] = (1 - tau) * w * st.z_grid[None, :]   # 労働所得
    yvec[NjW:] = st.psi0 * w                                     # 退職者は年金
    inc   = yvec[:, :, None] + r * a_grid                        # 労働所得+資本所得
    cons  = yvec[:, :, None] + (1 + r) * a_grid - a_pr
    s     = inc - cons                                           # 貯蓄額
    srate = s / inc                                              # 貯蓄率

    cfunJ = (cons * sd).sum(axis=(1, 2)) / sd.sum(axis=(1, 2))   # 年齢jの平均消費
    norm  = 1.0 / cfunJ[0]                                       # 正規化用

    # ───────────────────── グラフ① 資産ストック ─────────────────────
    age_vec = np.arange(1, Nj + 1) + 19
    plt.figure(figsize=(7, 4.5))
    plt.plot(age_vec, norm * afunJE[:, 1], 'k-',  lw=3, label='high')
    plt.plot(age_vec, norm * afunJE[:, 0], 'k-.', lw=3, label='low')
    plt.xlim(20, 80)
    plt.xlabel('Age'); plt.ylabel('Asset')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # ─────── 描く前に共通の下準備 ───────
    jc       = jc_plot -1
    srateAE  = srate[jc]   # (nz, na)
    sfunAE   = s[jc]       # (nz, na)
    # ───────────────────── Saving level ─────────────────────
    plt.figure(figsize=(7, 4.5))
    plt.plot(a_grid, sfunAE[1], 'k-',  lw=3, label='high')
    plt.plot(a_grid, sfunAE[0], 'k-.', lw=3, label='low')
    plt.axhline(0, ls=':', color='k')
    #plt.xlim(0, 20) # matlabではx軸を制限
    plt.xlabel('Asset'); plt.ylabel('Saving level')
    plt.legend(); plt.grid(True); plt.tight_layout()
    plt.show()

    # ───────────────────── Saving rate ─────────────────────
    plt.figure(figsize=(7, 4.5))
    plt.plot(a_grid, srateAE[1], 'k-',  lw=3, label='high')
    plt.plot(a_grid, srateAE[0], 'k-.', lw=3, label='low')
    plt.axhline(0, ls=':', color='k')
    #plt.xlim(0, 20) # matlabではx軸を制限
    plt.xlabel('Asset'); plt.ylabel('Saving rate')
    plt.legend(); plt.grid(True); plt.tight_layout()
    plt.show()

```

# 移行過程

```python
def solve_transition(st:Setting, K_SS0:float, ss0: Dict, K_SS1:float, ss1: Dict):
        """
        年金改革 (ψ0→ψ1) が導入されるときの移行経路を解く。

        返り値
        -------
        K_path  : ndarray (NT,)             … 資本ストックの移行パス
        V_init  : ndarray (Nj,nz,na)        … 改革１期目の価値関数 V_0(j,z,a)（厚生効果用、移行過程計算で必要なし）
        V_start : ndarray (NT,nz)           … 各コホート出生時 V_t(0,z,0)（厚生効果用、移行過程計算で必要なし）
        """

        # 初期化
        K_path = np.linspace(K_SS0, K_SS1, st.NT)
        K_new  = np.empty_like(K_path)

        V_start = np.empty((st.NT, st.nz))
        V_init  = None

        # backward と forward　で均衡条件の確認
        for it in range(st.max_iter_TR):
            v_next = ss1["V"]
            a_prime_idx_path = np.empty((st.NT, st.Nj, st.nz, st.na), dtype=np.int64)
            a_prime_path   = np.empty((st.NT, st.Nj, st.nz, st.na))

            # NT to 1 期
            for t in range(st.NT - 1, -1, -1):
                r = st.alpha * (K_path[t] / st.Lbar) ** (st.alpha - 1) - st.delta
                w = (1 - st.alpha) * (K_path[t] / st.Lbar) ** st.alpha
                rho, tau = st.psi_path[t], st.tau_path[t]
                v_curr, a_prime_idx, a_prime = bellman_backward(False,
                    v_next, r, w, tau, rho, st.beta, st.gamma, st.b,
                    st.a_grid, st.a_grid_sd, st.na, st.a_idx1, st.a_idx2, st.a_w1, st.a_w2,
                    st.z_grid, st.Pz, st.Nj, st.NjW, st.nz,
                )
                a_prime_idx_path[t] = a_prime_idx
                a_prime_path[t]   = a_prime
                v_next       = v_curr

                if t == 0:
                    V_init = v_curr.copy()
                V_start[t] = v_curr[0, :, 0]

            # 1 to NT 期
            sd = ss0["sd"]
            for t in range(st.NT - 1):
                K_new[t + 1] = aggregate_capital(a_prime_path[t], sd)
                sd = forward_measures(sd, a_prime_idx_path[t], st.a_idx1, st.a_idx2, st.a_w1, st.a_w2, st.Pz, st.sd_age, st.Nj, st.nz, st.na)

            # 誤差確認
            diff = np.max(np.abs(K_new - K_path))
            print(f"Iter {it:2d} | K error = {diff:1.2e}")
            if diff < st.tol_K_TR:
                break
            K_path[1:] += st.adjK_TR * (K_new[1:] - K_path[1:])
            K_new = np.copy(K_path)

        return K_path, V_init, V_start
```

```python
def plot_transition(st:Setting, K_path: np.ndarray):
        T = np.arange(st.NT)
        plt.figure(figsize=(8, 4))
        plt.plot(T, K_path/K_path[0], label="K")
        plt.xlabel("time")
        plt.ylabel("Capital")
        plt.legend()
        plt.tight_layout()
        plt.show()

        r = st.alpha * (K_path / st.Lbar) ** (st.alpha - 1) - st.delta
        plt.figure(figsize=(8, 4))
        plt.plot(T, r, label="r")
        plt.xlabel("time")
        plt.ylabel("r")
        plt.legend()
        plt.tight_layout()
        plt.show()

        return plt.gcf()

plot_transition(st, K_path)
```

```python
def plot_welfare(st, welf0_JE, welfTR, welf0, j40=21):
        T  = np.arange(st.NT) + 1
        nj = np.arange(st.Nj) + 20

        # 図1: 改革時点の年齢別厚生
        plt.figure()
        plt.plot(nj, welf0_JE[:,0], '--k', lw=2, label='low')
        plt.plot(nj, welf0_JE[:,1], '-k',  lw=2, label='high')
        plt.xlabel('Age'); plt.ylabel('CEV'); plt.grid(True)
        plt.legend(); plt.title('Initial cross-section welfare')

        # 図2: 改革後生まれた２０歳のCEVを、いつ生まれるかによって図示
        plt.figure()
        plt.plot(T, welfTR[:,0], '--k', lw=2, label='low')
        plt.plot(T, welfTR[:,1], '-k',  lw=2, label='high')
        plt.xlim(1,60); plt.xlabel('Birth-year cohort'); plt.ylabel('CEV')
        plt.legend(); plt.grid(True); plt.title('Welfare by cohort')

        # 図3: 改革時点４０歳の資産別CEV
        plt.figure()
        plt.plot(st.a_grid, welf0[j40, 0, :], '--k', lw=2, label='low')
        plt.plot(st.a_grid, welf0[j40, 1, :], '-k',  lw=2, label='high')
        plt.xlabel('Asset'); plt.ylabel('CEV'); plt.grid(True)
        plt.legend(); plt.title('Age 40 welfare by asset')

        plt.show()

# 年齢ごとの割引係数 β^(残余寿命) を列ベクトルで作成
# betaJ[j] = β^(Nj-j-1) + β^(Nj-j-2) + ... + 1
betaJ = np.array([(st.beta**np.arange(st.Nj-j)).sum() for j in range(st.Nj)])

# 改革時点 (t=0) の全世代・全資産に対する “消費補償変分” (CEV)
# CEV(j,z,a) = exp( ΔV(j,z,a) / Σ_t β^t ) - 1
welf0 = np.exp((V_init - V_SS0) / betaJ[:,None,None]) - 1

# 改革時点での「年齢×スキル」平均 CEV
# 資産分布 μ(j,z,a) で重み付けし、資産次元を積分
welf0_JE = (welf0 * ss0["sd"]).sum(axis=2) / ss0["sd"].sum(axis=2)

# 各コホート出生時 (t=0..NT-1) の平均 CEV
# V_start[t,z] はそのコホートの V(0,z,0)、
# V_SS0[0,z,0] は改革前 SS の出生価値
welfTR = np.exp((V_start - V_SS0[0,:,0]) / betaJ[0]) - 1


plot_welfare(st, welf0_JE, welfTR, welf0)
```
